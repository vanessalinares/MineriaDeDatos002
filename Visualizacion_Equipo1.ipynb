## - Scatter plot


```python
df.plot.scatter(x='Puntuacion', y='Generocidad')
```

Esta gráfica nos muestra que la Generocidad y la Puntuacion de los paises estan muy poco relacionados. <br> <br>
Esto tambien lo pudimos ver con la covarianza (esto se puede encontrar justo al final del apartado de Estadistica Báisica) donde nos muestra que la covarianza entre la Puntuacion de un país y su Generocidad es de <b>0.008040</b>. Esta es una covarianza positiva, lo cual nos dice que tienen una relación directa, sin embargo es un numero muy pequeño, lo cual nos dice que es una <b>relacion directa débil</b>.
<br> <br> Tambien podemos basarnos en la correlación mostrada entre ambas variables (esto también se puede encontrar justo al final del apartado de Estadistica Básica) donde nos muestra que la correlacion entre la Puntuacion de un país y su Generocidad es de <b>0.075824</b>. Al ser un valor positivo denota relacion directa, pero al ser un valor muy cercano al 0 muestra una <b>relación directa debil</b>.

```python
df.plot.scatter(x='Puntuacion', y='Apoyo_Social')
```

Al analizar esta gráfica visualmente, podemos darnos cuenta de que el Apoyo Social en un país muestra una relacion lineal positiva con la Puntuacion del país.
<br><br>
Esto tambien lo pudimos ver con la covarianza (esto se puede encontrar justo al final del apartado de Estadistica Báisica) donde nos muestra que la covarianza Entre la Puntuacion de un país y su Generocidad es de <b>0.258788</b>. Esta es una covarianza positiva, lo cual nos dice que tienen una relación directa, sin embargo es un numero muy pequeño, lo cual nos dice que es una <b>relacion directa débil</b>.
<br> <br>
Tambien podemos basarnos en la correlación mostrada entre ambas variables (esto también se puede encontrar justo al final del apartado de Estadistica Básica) donde nos muestra que la correlacion entre la Puntuacion de un país y su Apoyo Social es de <b>0.777058</b>. Al ser un valor positivo denota relacion directa, asi mismo es un valor cercano al 1 (aunque no tan cercano, pero es mas cercano a 1 que a 0) muestra una <b>relación directa fuerte </b>.
<br> <br>
Estas dos conclusiones se contradicen un poco, ya que la conclusion respecto a la varianza nos dice que tienen una <b>relacion directa débil</b> pero la conclusion de la correlación nos dice que tienen una <b>relación directa fuerte </b>

Esta gráfica nos muestra la densidad de los países de acuerdo a su puntuación, con la ténica de KDE.

```python
sns.kdeplot(df['Esperanza_de_vida'],df['Puntuacion'].dropna())
```

La grafica anterior es una gráfica de distancias de una distribución bivariante, donde la variable x es la esperanza de vida y la variable y es la puntuacion

```python
sns.distplot(df['Puntuacion'], bins=10, kde=True)
```

El gráfico distplot representa la distrubucion de la puntuación, básicamente es un histograma.

```python
sns.jointplot(x='Esperanza_de_vida', y='Puntuacion', data=df[df['Puntuacion'] < 10])
```

Esta gráfica nos mustra tanto un scatter plot como un histograma para cada una de las variables, en este caso x=Esperanza de vida y y=Puntuacion. Como podemos ver en el Scatter Plot, nuestras variables muestran una relación lineal positiva. Con los histogramas nos podemos dar cuenta de que la puntuación mas frecuente en los países es de 5.5 y la Esperanza de vida mas frecuente en los países es de 0.8

```python
sns.jointplot(x='Esperanza_de_vida', y='Puntuacion', data=df[df['Puntuacion'] < 10], kind='hex', 
              gridsize=20)
```

Ahora, esta gráfica nos mustra tanto un hexplot como un histograma para cada una de las variables, estamos usando las mismas variables que en el grafico anterior (x=Esperanza de vida y y=Puntuacion). Como podemos ver en el hexplot, se hacen varios clusters.

## - Box plots

No es posible hacer un box plot coherente con nuestra base de datos. La razón es que sabemos que los Boxplot tabajan con variables de intervalo y con variables nominales con una gran candidad de posibles valores. El "problema" con nuestra base de datos es que la mayoria (si no es que todos) son datos únicos. Entonces, al momento de hacer un box plot la gráfica no tiene sentido, como se puede ver a continuación:

```python
dfTF = df[df.Pais_o_region.isin(df.Pais_o_region.value_counts().head(5).index)]

sns.boxplot(
    x='Puntuacion',
    data=dfTF
)
```

## - Pairplot

```python
sns.pairplot(df[['Puntuacion', 'PIB_per_capita', 'Apoyo_Social', 'Esperanza_de_vida', 'Libertad','Generocidad', 'Percepción_de_corrupcion']])
```

Esta gráfica pairplot nos permite ver los scatterplots entre cada una de nuestras columnas a analizar, lo cual hace más sencillo ver cual es la relación entre ellas de una manera gráfica.
